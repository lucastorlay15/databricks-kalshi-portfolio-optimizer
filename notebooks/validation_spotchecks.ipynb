```python
%sql
SELECT
  ingest_mode,
  COUNT(*) AS trade_rows,
  COUNT(DISTINCT trade_id) AS distinct_trade_ids,
  MIN(trade_created_time_utc) AS first_trade_time,
  MAX(trade_created_time_utc) AS last_trade_time,
  MIN(yes_price_dollars) AS min_yes_price,
  MAX(yes_price_dollars) AS max_yes_price,
  SUM(contracts_traded) AS total_contracts
FROM databricks_realtime_optimization.cfb_analytics.bronze_kalshi_cfb_trades
GROUP BY ingest_mode;
```
